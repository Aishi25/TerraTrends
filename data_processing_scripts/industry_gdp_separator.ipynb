```python
def process_df_gdp(df: pd.DataFrame, sector_name_dict: dict[str, str]) -> pd.DataFrame:
    gdp_cols = sector_name_dict.keys()
    remaining_cols = set(df.columns) - set(gdp_cols)
    new_df = pd.melt(
        frame=df,
        value_vars=gdp_cols,
        id_vars=remaining_cols,
        value_name="Sector_GDP",
        var_name="Sector",
	)

    new_df["Sector"] = new_df["Sector"].map(sector_name_dict)

    return new_df
```
