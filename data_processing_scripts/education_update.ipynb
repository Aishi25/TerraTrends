```python
import pandas as pd
import os

INPUT_DATA_PATH = "../data/input"
OUTPUT_DATA_PATH = "../data/output"
NEW_DATA_FILE_NAME = "2006-2010_edu_data.csv"
OLD_DATA_FILE_NAME = "merged_data.csv"
OUTPUT_FILE_NAME = "final_merged_data.csv.csv"

TARGET_YEARS = [2006, 2007, 2008, 2009]
```

```python
def load_csv_data(input_path: str, file_name: str) -> pd.DataFrame:
    data_path = os.path.join(input_path, file_name)
    return pd.read_csv(data_path)

def combine_df_education_data(old_df: pd.DataFrame, new_df: pd.DataFrame) -> pd.DataFrame:
    merged = old_df.merge(
        new_df[['County', 'Year', 'Education']],
        on=['County', 'Year'],
        how='left',
        suffixes=('', '_new')
    )

    merged['TOT_POP'] = merged['TOT_POP'].fillna(merged['Population'])

    merged = merged.drop(columns=['Population'])
    return merged

def change_data_format(df: pd.DataFrame, dropped_years: list[int], target_years: list[int], id_columns: list[str]) -> pd.DataFrame:
    new_df = df.drop(dropped_years, axis=1)
    return pd.melt(
        frame=new_df,
        id_vars=id_columns,
        value_vars=target_years,
        var_name="Year",
        value_name="Population",
    ).sort_values(["County", "Year"])

def save_df_to_csv(df: pd.DataFrame, output_path: str, file_name: str) -> None:
    data_path = os.path.join(output_path, file_name)
    df.to_csv(data_path, index=False)
```
